```python
import pandas as pd
import matplotlib.pyplot as plt

initialValues = [0.001,0,0.1,0.5,1]

# Initialize the figure and its subplots
plt.figure(figsize=(24, 12))

# Subplots for GD
# Plot for Time vs Loss for GD
plt.subplot(3, 2, 1)
for i in range(1, 6):
    df_gd = pd.read_csv(f'./GD/loss_time_epocks/gd_performance{i}.csv')
    plt.plot(df_gd['Time'], df_gd['Loss'], marker='o', label=f'GD initial value {initialValues[i-1]}')
plt.title('GD: Loss vs Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Plot for Loss per Epoch for GD
plt.subplot(3, 2, 2)
for i in range(1, 6):
    df_gd = pd.read_csv(f'./GD/loss_time_epocks/gd_performance{i}.csv')
    plt.plot(df_gd['Epoch'], df_gd['Loss'], marker='o', label=f'GD initial value {initialValues[i-1]}')
plt.title('GD: Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Subplots for SGD
# Plot for Time vs Loss for SGD
plt.subplot(3, 2, 3)
for i in range(1, 6):
    df_sgd = pd.read_csv(f'./SGD/loss_time_epocks/gd_performance{i}.csv')
    plt.plot(df_sgd['Time'], df_sgd['Loss'], marker='o', label=f'SGD initial value {initialValues[i-1]}')
plt.title('SGD: Loss vs Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Plot for Loss per Epoch for SGD
plt.subplot(3, 2, 4)
for i in range(1, 6):
    df_sgd = pd.read_csv(f'./SGD/loss_time_epocks/gd_performance{i}.csv')
    plt.plot(df_sgd['Epoch'], df_sgd['Loss'], marker='o', label=f'SGD initial value {initialValues[i-1]}')
plt.title('SGD: Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()


# Subplots for ADAM
# Plot for Time vs Loss for ADAM
plt.subplot(3, 2, 5)
for i in range(1, 6):
    df_sgd = pd.read_csv(f'./ADAM/loss_time_epocks/gd_performance{i}.csv')
    plt.plot(df_sgd['Time'], df_sgd['Loss'], marker='o', label=f'ADAM initial value {initialValues[i-1]}')
plt.title('ADAM: Loss vs Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Plot for Loss per Epoch for ADAM
plt.subplot(3, 2, 6)
for i in range(1, 6):
    df_sgd = pd.read_csv(f'./ADAM/loss_time_epocks/gd_performance{i}.csv')
    plt.plot(df_sgd['Epoch'], df_sgd['Loss'], marker='o', label=f'ADAM initial value {initialValues[i-1]}')
plt.title('ADAM: Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()




# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Initial values, assumed to be used for reference or annotation
initialValues = [0.001,0,0.1,0.5,1]

# Colors for each set of weights
colors = ['red', 'green', 'blue', 'purple', 'orange']

# Setting up the plot - adjusting for three sets
plt.figure(figsize=(24, 12))  # Adjusted figure size for 3 plots

# Plotting the first set (GD)
plt.subplot(1, 3, 1)
for i in range(1, 6):
    weights_df = pd.read_csv(fr'./GD/weigts/weights_per_epoch{i}.csv')
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('First Set: t-SNE Visualization of Weights (GD)')
plt.legend()
plt.xlim(-50, 50)
plt.ylim(-50, 50)

# Plotting the second set (SGD)
plt.subplot(1, 3, 2)
for i in range(1, 6):
    weights_df = pd.read_csv(f'./SGD/weigts/weights_per_epoch{i}.csv')
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('Second Set: t-SNE Visualization of Weights (SGD)')
plt.legend()
plt.xlim(-50, 50)
plt.ylim(-50, 50)

# Plotting the third set (ADAM)
plt.subplot(1, 3, 3)
for i in range(1, 6):
    weights_df = pd.read_csv(f'./ADAM/weigts/weights_per_epoch{i}.csv')
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('Third Set: t-SNE Visualization of Weights (ADAM)')
plt.legend()
plt.xlim(-50, 50)
plt.ylim(-50, 50)

plt.show()
```
