```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Initial values, assumed to be used for reference or annotation
initialValues = [0.001,0,0.1,0.5,1]

# Colors for each set of weights
colors = ['red', 'green', 'blue', 'purple', 'orange']

# Setting up the plot
plt.figure(figsize=(12, 8))

for i in range(1, 2):
    # Step 1: Read the CSV file
    weights_df = pd.read_csv(f'C:\\Users\\BTK\\Desktop\\diferansiyel\\ADAM\\weigts\\weights_per_epoch{i}.csv')  # Adjust the path as necessary
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column

    # Find the number of samples
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)  # Set perplexity to be less than the number of samples

    # Step 2: Apply t-SNE with adjusted parameters
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)

    # Step 3: Visualize with Matplotlib
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

# Adding labels, title, and legend
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('t-SNE Visualization of Weights from Different Files')
plt.legend()

# Setting the x and y axis limits to be centered around the clusters
plt.xlim(-50, 50)
plt.ylim(-50, 50)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Initial values, assumed to be used for reference or annotation
initialValues = [0.001,0,0.1,0.5,1]

# Colors for each set of weights
colors = ['red', 'green', 'blue', 'purple', 'orange']

# Setting up the plot - adjusting for three sets
plt.figure(figsize=(24, 12))  # Adjusted figure size for 3 plots

# Plotting the first set (GD)
plt.subplot(1, 3, 1)
for i in range(1, 6):
    weights_df = pd.read_csv(f'C:\\Users\\BTK\\Desktop\\diferansiyel\\GD\\weigts\\weights_per_epoch{i}.csv')
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('First Set: t-SNE Visualization of Weights (GD)')
plt.legend()
plt.xlim(-50, 50)
plt.ylim(-50, 50)

# Plotting the second set (SGD)
plt.subplot(1, 3, 2)
for i in range(1, 6):
    weights_df = pd.read_csv(f'C:\\Users\\BTK\\Desktop\\diferansiyel\\SGD\\weigts\\weights_per_epoch{i}.csv')
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('Second Set: t-SNE Visualization of Weights (SGD)')
plt.legend()
plt.xlim(-50, 50)
plt.ylim(-50, 50)

# Plotting the third set (ADAM)
plt.subplot(1, 3, 3)
for i in range(1, 6):
    weights_df = pd.read_csv(f'C:\\Users\\BTK\\Desktop\\diferansiyel\\ADAM\\weigts\\weights_per_epoch{i}.csv')
    weights = weights_df.iloc[:, 1:]  # Exclude the epoch column
    num_samples = weights.shape[0]
    perplexity_value = min(30, num_samples - 1)
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity_value)
    weights_reduced = tsne.fit_transform(weights)
    plt.scatter(weights_reduced[:, 0], weights_reduced[:, 1], c=colors[i-1], label=f'Initial Value: {initialValues[i-1]}')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('Third Set: t-SNE Visualization of Weights (ADAM)')
plt.legend()
plt.xlim(-50, 50)
plt.ylim(-50, 50)

plt.show()
```
